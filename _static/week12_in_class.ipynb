## bar graphs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
country=["italy","us","canada","india"]
pop=[60,334,38,1399]
```

```python
plt.bar(country,pop,color=(1,0,1),width=0.9)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(range(len(pop)),pop,color=(1,0,1),width=0.5,tick_label=country)
plt.xticks(rotation=45)
plt.show()
```

## function behavior

```python
x=np.arange(1,100)
y=x**2
```

```python
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(np.log(x),np.log(y))
```

```python
x=np.arange(1,100)
y=np.exp(-x)
```

```python
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(x,np.log(y)) #semi-log y
```

```python
x=np.arange(1,100)
y=np.log(x)
```

```python
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(np.log(x),y) #semi-log x
```
