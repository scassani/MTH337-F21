### Step 0

Given a convergent series (either p-series or geometric)

$p=2$, $\sum_{n=1}^\infty \frac{1}{n^2}$

* find radius, center (0,0) of the black circle. How to find `R`?  
* $A=\sum_{n=1}^\infty \frac{1}{n^p}=\zeta(2)=1.6...$

* $R=\sqrt{\frac{A}{\pi}}$

 

```python
from scipy.special import zetac
```

```python
1+zetac(2) # sum without n=1 term
```

```python
A=1+zetac(2)
R=(A/pi)**0.5
R
```

```python
from matplotlib.patches import Circle
```

```python
ax1=plt.subplot(1,1,1)
ax1.add_patch( Circle([0,0],R,fc='k',zorder=1) )
ax1.add_patch( Circle([0.5,0.5],0.1,fc='y',zorder=2) )
ax1.add_patch( Circle([-0.25,-0.25],0.05,fc='c',zorder=2) )
plt.axis('equal')
plt.xlim([-1.1*R,1.1*R])
plt.ylim([-1.1*R,1.1*R])
plt.show()
```
