# colors by RGB

```python
plt.plot(0,0,'s',color=(0,1,0),markersize=40) # big green square
```

```python
w=0.5
h=(3**0.5)/2
m=40

plt.figure(figsize=(4,4))
ax=plt.subplot(1,1,1)
plt.plot(0,h,'*',markersize=m,color=(1,0,0)) # red
plt.plot(w,0,'*',markersize=m,color=(0,1,0)) # green
plt.plot(-w,0,'*',markersize=m,color=(0,0,1)) # blue
plt.plot(w/2,h/2,'*',markersize=m,color=(1,1,0)) # red + green
plt.plot(-w/2,h/2,'*',markersize=m,color=(1,0,1)) # red + blue
plt.plot(0,0,'*',markersize=m,color=(0,1,1)) # green + blue
plt.plot(0,h/2,'*',markersize=m,color=(1,1,1)) # red + green + blue
ax.set_facecolor((0,0,0)) # black background
```

```python
a=np.random.rand(4,4,3)
plt.imshow(a);
```

```python
a[0,0,:] # amout of red-green-blue in firstpixel (position [0,0])
```

how do I plot in red scale??

```python
a[:,:,0] # red component of the image
```

```python
plt.imshow(a[:,:,0]) # this is not working to plot the red scale
```

```python
# plot with red-scale only
r=a.copy()

r[:,:,1:3]=0 # zero-out green and blue

plt.imshow(r)
```

```python
# plot with green-scale only
g=a.copy()

g[:,:,0]=0 # zero-out red
g[:,:,2]=0 # zero-out blue

plt.imshow(g)
```
